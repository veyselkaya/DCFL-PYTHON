```python
#kullanıcı adı ve şifre konrtolü yapan program

k_adi = input("kullanıcı adını gir:")
sifre = input("şifreni gir: ")

if k_adi=="kedi" and sifre=="1234":
    print("bilgilerin doğru...hoş geldin")
elif k_adi=="kedi" and sifre!="1234":
    print("şifren yanlış...bir daha dene")
elif k_adi!="kedi" and sifre=="1234":
    print("kuıllanıcı adın yanlış...bir daha dene") 
else:
    print("girdiğin iki bilgi de yanlış.....!")
```
