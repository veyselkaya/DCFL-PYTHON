```python
###### Çok fonksiyonlu hesap makinesi

print("""*******************
Hesap makinesi
toplama için 1
Çıkarma için 2
Bölme için 3
çarpma için 4
üs alma için de 5 e basınız
""")

islem = input("İşlemini seç")

sayi1 = int(input("Birinci sayıyı gir:"))
sayi2 = int(input("İkinci sayıyı gir:"))

if islem =="1":
    print("Toplama işlemi seçildi")
    print("{} sayıs ile {} nın toplamı {} dır.".format(sayi1, sayi2, sayi1+sayi2))
elif islem =="2":
    print("Çıkartma işlemi seçildi")
    print("{} sayıs ile {} nın farkı {} dır.".format(sayi1, sayi2, sayi1-sayi2))
elif islem =="3":
    print("Bölme işlemi seçildi")
    print("{} sayıs ile {} nın bölümü {} dır.".format(sayi1, sayi2, sayi1/sayi2))
elif islem =="4":
    print("Çarpma işlemi seçildi")
    print("{} sayıs ile {} nın çarpımı {} dır.".format(sayi1, sayi2, sayi1*sayi2))
elif islem =="5":
    print("Üs alma işlemi seçildi")
    print("{} sayısının {} kuvetti {} dır.".format(sayi1, sayi2, sayi1**sayi2))
else:
    print("Lütfen geçerli bir işlem seç!")
```
