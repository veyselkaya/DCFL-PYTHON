```python
#geometri formülleri programı

print("""Formülünü seç
karenin alanı için 1
üçgenin alanı için 2
.....
""")

formul= input("Formülünü seç:")

if formul =="1":
    print("Karenin alanını bulma programı seçildi:")
    a = int(input("karenin kenarını gir:"))
    print("Girdiğin kenar uzunluığu {} dır. Karenin alanı {} dır".format(a,a**a))
    
elif formul =="2":
    print("Üçgenin alanını bulma programı seçildi:")
    taban = int(input("Üçgenin tabanını gir:"))
    yukseklik = int(input("Üçgenin Yüksekliğini gir:"))
    print("Girdiğin taban uzunluğu {} dır, Yüksekliği {}. Üçgenin  alanı {} dır".format(taban, yukseklik, (taban*yukseklik)/2 ))
else:
    print("Lütfen geeçrli bir işlem seçiniz")
```
